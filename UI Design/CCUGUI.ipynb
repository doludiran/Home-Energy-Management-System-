```python
import numpy as np
import matplotlib.pyplot as plt

men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                color='SkyBlue', label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                color='IndianRed', label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')


autolabel(rects1, "left")
autolabel(rects2, "right")

plt.show()
```

```python
from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')

from kivy.garden.matplotlib.backend_kivyagg import FigureCanvasKivyAgg
from kivy.app import App
from kivy.uix.boxlayout import BoxLayout
import matplotlib.pyplot as plt
import numpy as np
import sqlite3


width = 1  # the width of the bars

fig, ax = plt.subplots()


TimeOfUse = sqlite3.connect('ToU')
cursor = TimeOfUse.cursor()


cursor.execute('''SELECT time, price, timeofuse FROM ToU''')
tou_data = cursor.fetchall()
xlabel =[]
for row in tou_data:
    if row[2] == 'offpeak':
        ax.bar(row[0], row[1], width,
                    color='LightGreen',label='Off Peak')
    else:
        ax.bar(row[0], row[1], width,
                    color='IndianRed',label='On Peak')
    xlabel.append(row[0])
    
ind=np.arange(len(xlabel))

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Price per kWh')
ax.set_frame_on(False)
ax.set_title('Time of Use Schedule')
ax.set_xticks(ind)
ax.set_xticklabels(xlabel)
ax.autoscale(enable=True, axis='both', tight=None)
ax.legend()


class MyApp(App):

    def build(self):
        box = BoxLayout()
        box.add_widget(FigureCanvasKivyAgg(plt.gcf()))
        return box

MyApp().run()
```
