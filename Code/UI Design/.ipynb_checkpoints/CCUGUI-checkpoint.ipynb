# Kivy Tutorial 1
Tutorial Source: 
Kivy Crash course by Alexander Taylor: https://www.youtube.com/watch?v=F7UKmK9eQLY&t=6s

Objective:
Get familiar with python and the Kivy framework, kivy App and kivy widgets


```python
#configure the graphics display window
from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')

#import the neccesary modules for the python app
from kivy.app import App
from kivy.graphics import Color, Rectangle
from kivy.uix.scatter import Scatter
from kivy.uix.label import Label
from kivy.uix.floatlayout import FloatLayout
from kivy.uix.textinput import TextInput
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.button import Button

#Implementation of the user interface class. 
class GraphicUI(BoxLayout):
    pass

#The Kivy App class
class CCU(App):
    def build(self):
        btn =Button(text='ENEL 400', font_size=30, size_hint=(.1,.1))
        b = BoxLayout(orientation='vertical')
        t = TextInput(font_size=150,
                     size_hint_y=None,
                     height =200,
                     text='default')
        
        f = FloatLayout()
        s = Scatter()
        l = Label(text="Hello! ENEL 400",
                  font_size=150)
        
        t.bind(text=l.setter('text'))
        
        f.add_widget(s)
        s.add_widget(l)
        
        b.add_widget(t)
        b.add_widget(f)
        b.add_widget(btn)
        return b

if __name__ == "__main__":
    CCU().run()
```

# ENEL 400 IO Characterization

A simple UI with some text and a button to turn on a LED with the Raspberry PI 

```python
from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')

from kivy.app import App
from kivy.graphics import Color, Rectangle
from kivy.uix.label import Label
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.anchorlayout import AnchorLayout
from kivy.uix.button import Button

class CCUgraphicUI(App):
    def build(self):
        btn =Button(text='LED', font_size=40, size_hint=(.15,.15))
        l = Label(text="Hello! ENEL 400",
                  font_size=150)
        b = BoxLayout(orientation='vertical')
        a = AnchorLayout(anchor_x='center')
        
        a.add_widget(btn)
        b.add_widget(l)
        b.add_widget(a)
        return b

if __name__ == "__main__":
    CCUgraphicUI().run()
```

# Sample code from matplotlib.org

- plot of a simple bar chart using matplotlib modules

```python
import numpy as np
import matplotlib.pyplot as plt

men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                color='SkyBlue', label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                color='IndianRed', label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')


autolabel(rects1, "left")
autolabel(rects2, "right")

plt.show()
```

# Time of Use Plot

Objective:
- Generate a time of use plot from database data

```python
from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')

from kivy.garden.matplotlib.backend_kivyagg import FigureCanvasKivyAgg
from kivy.app import App
from kivy.uix.boxlayout import BoxLayout
import matplotlib.pyplot as plt
import numpy as np
import sqlite3


width = 1  # the width of the bars

fig, ax = plt.subplots()


TimeOfUse = sqlite3.connect('ToU')
cursor = TimeOfUse.cursor()


cursor.execute('''SELECT time, price, timeofuse FROM ToU''')
tou_data = cursor.fetchall()
xlabel =[]
for row in tou_data:
    if row[2] == 'offpeak':
        ax.bar(row[0], row[1], width,
                    color='LightGreen',label='Off Peak')
    else:
        ax.bar(row[0], row[1], width,
                    color='IndianRed',label='On Peak')
    xlabel.append(row[0])
    
ind=np.arange(len(xlabel))

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Price per kWh')
ax.set_frame_on(False)
ax.set_title('Time of Use Schedule')
ax.set_xticks(ind)
ax.set_xticklabels(xlabel)
ax.autoscale(enable=True, axis='both', tight=None)
ax.legend()


class MyApp(App):

    def build(self):
        box = BoxLayout()
        box.add_widget(FigureCanvasKivyAgg(plt.gcf()))
        return box

MyApp().run()
```

```python
dotKV = '''\
<CCUGUI>:
    id: manager
    Screen:
        name: 'Home'
        GridLayout:
            rows: 3
            canvas:
                Color:
                    rgba: 1, 1, 1, 1
                Rectangle:
                    size: root.width, root.height
                    pos: 0, 0
            GridLayout:
                cols: 2
                size_hint: .2 , .2
                BoxLayout:
                    orientation: 'vertical'
                    size_hint: 1 , .2
                    Label:
                        text: '4/1/2019'
                        color: 0, 0, 0, 1
                    Label:
                        text: '12:00PM'
                        color: 0, 0, 0, 1
                BoxLayout:
                    orientation: 'vertical'
                    Button:
                        id: 'Home'
                        size_hint: .15, .2
                        on_press: manager.current= 'Menu'
                        Image:
                            source: 'icon.png'
                            center_y: self.parent.center_y 
                            center_x: self.parent.center_x
                            size: 250, 250
                            allow_strech: True
            BoxLayout:
                orientation: 'vertical'
                canvas:
                    Color:
                        rgba: 0, 0, 0, 0
                    Rectangle:
                        size: self.width, self.height
                size_hint: .3, .3
                Label:
                Label:
            BoxLayout:
                id: touplot
                orientation: 'horizontal'
                size_hint: .5, .5
                
    Screen:
        name: 'Menu'
        BoxLayout:
            canvas:
                Color:
                    rgba: 1, 1, 1, 1
                Rectangle:
                    size: root.width, root.height
                    pos: 0, 0
            AnchorLayout:
                orientation: 'horizontal'
                anchor_x: 'center'
                GridLayout:
                    cols: 2
                    size_hint: .31, .41
                    Button:
                        id: 'Therm'
                        on_press: manager.current= 'Thermostat'
                        size_hint: .15, .15
                        Image:
                            source: 'thermostat.png'
                            center_y: self.parent.center_y 
                            center_x: self.parent.center_x
                            size: 250, 250
                            allow_strech: True
                    Button:
                        id: 'EV'
                        on_press: manager.current= 'Car'
                        size_hint: .15, .15
                        Image:
                            source: 'car.png'
                            center_y: self.parent.center_y 
                            center_x: self.parent.center_x
                            size: 250, 250
                            allow_strech: True
                    Button:
                        id: 'Water'
                        size_hint: .15, .15
                        Image:
                            source: 'water.png'
                            center_y: self.parent.center_y 
                            center_x: self.parent.center_x
                            size: 250, 250
                            allow_strech: True
                    Button:
                        id: 'x10'
                        size_hint: .15, .15
                        Image:
                            source: 'x10.png'
                            center_y: self.parent.center_y 
                            center_x: self.parent.center_x
                            size: 250, 250
                            allow_strech: True
                
    Screen:
        name: 'Car'
        BoxLayout:
            canvas:
                Color:
                    rgba: 1, 1, 1, 1
                Rectangle:
                    size: root.width, root.height
                    pos: 0, 0
            BoxLayout:
                orientation: 'horizontal'
                size_hint: 0.1, 0.1
                Button:
                    id: 'schedule'
                    text: 'Schedule'
                    font_size: 40
                    size: 15, 15
                    background_color: 0, 2, 0, 0.5
                Button:
                    id: 'usage'
                    text: 'Usage'
                    font_size: 40
                    size: 15, 15
                    background_color: 0, 2, 0, 0.5

            
    Screen:
        name: 'Thermostat'
        BoxLayout:
            canvas:
                Color:
                    rgba: 1, 1, 1, 1
                Rectangle:
                    size: root.width, root.height
                    pos: 0, 0
            GridLayout:
                id: thermostat
                rows: 2
                colour: (2, 0, 0, 1)
                btn_colour: (2.5, 0, 0, 1)
                BoxLayout:
                    orientation: 'vertical'
                    Label:
                        id: temperature
                        text: "Thermostat"
                        color: self.parent.parent.colour
                        font_size: 50
                    Button:
                        id: increase
                        background_color: 0, 0, 0, 0
                        canvas:
                            Color: 
                                rgba: self.parent.parent.colour
                            Triangle:
                                points: [self.center_x - 70, self.center_y - 20,self.center_x, self.center_y + 80,self.center_x + 70, self.center_y - 20]
                        
                    Label:
                        text: "25"
                        color: self.parent.parent.colour
                        font_size: 300
                    Button:
                        id: 'decrease'
                        background_color: 0, 0, 0, 0
                        size: 10, 10
                        canvas:
                            Color:
                                rgba: self.parent.parent.colour
                            Triangle:
                                points: [self.center_x - 70, self.center_y + 20,self.center_x, self.center_y - 80,self.center_x + 70, self.center_y + 20]
                        
                    Label:
                        id: mode
                        text: "Heating"
                        color: self.parent.parent.colour
                        font_size: 30
                        
                            
                BoxLayout:
                    orientation: 'horizontal'
                    size_hint: 0.1, 0.1
                    Button:
                        id: heat
                        text: 'Heat'
                        font_size: 40
                        background_color: self.parent.parent.btn_colour
                        on_press: root.heat_mode()
                    Button:
                        id: cool
                        text: 'Cool'
                        font_size: 40
                        size: 10, 10
                        background_color: self.parent.parent.btn_colour
                        on_press: root.cool_mode()
                    Button:
                        id: eco
                        text: 'Schedule'
                        font_size: 40
                        size: 10, 10
                        background_color: self.parent.parent.btn_colour  
                    Button:
                        id: usage
                        text: 'History'
                        font_size: 40
                        size: 10, 10
                        background_color: self.parent.parent.btn_colour
    
    
        

\
'''
from kivy.lang import Builder
Builder.load_string(dotKV)
from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')

from kivy.app import App
from kivy.garden.matplotlib.backend_kivyagg import FigureCanvasKivyAgg
from kivy.uix.screenmanager import ScreenManager, Screen
from kivy.graphics import Color, Rectangle
from kivy.uix.behaviors import ButtonBehavior
from kivy.uix.image import Image
from kivy.uix.label import Label
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.anchorlayout import AnchorLayout
from kivy.uix.gridlayout import GridLayout
from kivy.uix.button import Button
import matplotlib.pyplot as plt
import numpy as np
import sqlite3

class CCUGUI(ScreenManager):
    def __init__(self):
        ScreenManager.__init__(self)
        self.data_graphs()
        self.ids.touplot.add_widget(FigureCanvasKivyAgg(plt.gcf()))
 
        
    def cool_mode(self, *args):
        colour = [0,0.6588,1,1]
        btn_colour = [0, 1.3176, 2, 1]
        self.ids.thermostat.colour = colour
        self.ids.thermostat.btn_colour = btn_colour
        self.ids.mode.text = 'Cooling'
        
    def heat_mode(self, *args):
        colour = [2,0,0,1]
        btn_colour = [2.5,0,0,1]
        self.ids.thermostat.colour = colour
        self.ids.thermostat.btn_colour = btn_colour
        self.ids.mode.text = 'Heating'
        
    def data_graphs(self, *args):
        width = 1  # the width of the bars

        fig, (ax0, ax1) = plt.subplots(ncols=2)


        TimeOfUse = sqlite3.connect('ToU')
        cursor = TimeOfUse.cursor()
        
        cmap = plt.get_cmap("tab20c")
        size = 0.4
        vals = np.array([[60., 32.], [37., 40.], [29., 10.]])
        outer_colors = cmap(np.arange(3)*4)

        cursor.execute('''SELECT time, price, timeofuse FROM ToU''')
        tou_data = cursor.fetchall()
        xlabel =[]
        for row in tou_data:
            if row[2] == 'offpeak':
                ax0.bar(row[0], row[1], width,
                            color='LightGreen',label='Off Peak')
            else:
                ax0.bar(row[0], row[1], width,
                            color='IndianRed',label='On Peak')
            xlabel.append(row[0])

        ind=np.arange(len(xlabel))

        # Add some text for labels, title and custom x-axis tick labels, etc.
        ax0.set_ylabel('Price per kWh')
        ax0.set_frame_on(False)
        ax0.set_title('Time of Use Schedule')
        ax0.set_xticks(ind)
        ax0.set_xticklabels(xlabel)
        ax0.autoscale(enable=True, axis='both', tight=None)
        ax0.legend()
        
        ax1.pie(vals.sum(axis=1), radius=1.2, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
        ax1.set(aspect="equal", title='Total Usage')
        

class CCUgraphicUI(App):
    def build(self):
        return CCUGUI()

if __name__ == "__main__":
    CCUgraphicUI().run()
```
